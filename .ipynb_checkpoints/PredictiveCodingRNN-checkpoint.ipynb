```python
import numpy as np
from matplotlib import pyplot as plt
```

# Input Data

```python
input_size = 90

rawWave = np.sin(np.radians(8 * np.arange(0,input_size,dtype=float)))
noise = np.random.normal(0,1e-1,input_size)
drift = np.arange(0., float(input_size)/1000, 0.001, dtype=float)

signal = rawWave + noise + drift
plt.plot(np.arange(0,input_size),signal)
plt.show()

input_data = np.zeros((input_size,input_size))
np.fill_diagonal(input_data, 1)
```

# Model

```python
# Modified from https://github.com/KTAswathi/IPB_2023/blob/main/IPB_lab3_BPTT.ipynb

import numpy as np
from numpy.random import randn

class RNN:
    # A Vanilla Recurrent Neural Network.

    def __init__(self, input_size, output_size, hidden_size=64):
        # Weights
        self.Whh = randn(hidden_size, hidden_size) / 1000
        self.Wxh = randn(hidden_size, input_size) / 1000
        self.Why = randn(output_size, hidden_size) / 1000
        self.Wfeedback = randn(hidden_size, output_size) / 1000
        
        # Biases
        self.bh = np.zeros((hidden_size, 1))
        self.by = np.zeros((output_size, 1))
    
    def forward(self, inputs, pred_weight, lr, gts):
        '''
        Perform a forward pass of the RNN using the given inputs.
        Returns the final output and hidden state.
        - inputs is an array of one-hot vectors with shape (input_size, 1).
        '''
        h = np.zeros((self.Whh.shape[0], 1))
        self.last_inputs = inputs
        #self.last_hs = { 0: h }
        self.last_pred = { 0: h }

        loss = []
        
        # Perform each step of the RNN
        for i, x in enumerate(inputs):
            #h = np.tanh(self.Wxh @ x + self.Whh @ h + self.bh)
            #self.last_hs[i + 1] = h

            h = np.tanh(self.Wxh @ x + self.Whh @ h + self.bh)
            
            # Compute prediction and prediction error signal
            y = self.Why @ h + self.by
            output_error = gts[i] - y

            h_pred = h + pred_weight * self.Wfeedback @ output_error # Might need to do something to dot these
            pred_error = h_pred - h
            self.last_pred[i + 1] = h_pred

            self.Whh = self.Whh + lr * pred_error * np.transpose(np.tanh(self.last_pred[i]))
            self.Wfeedback = self.Wfeedback + lr * np.tanh(h) @ np.transpose(output_error)

            loss.append(np.abs(np.mean(output_error)))
            
        
        return np.mean(np.abs(output_error))
        # Compute the output
        # y = self.Why @ h + self.by
        # return y

rnn = RNN(input_size,1,hidden_size=100)
loss_curve = []
for i in range(10):
    loss_curve.append(rnn.forward(input_data, 0.01, 0.1, signal))

plt.plot(np.arange(10),loss_curve) 
```
